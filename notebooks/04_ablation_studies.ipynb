This notebook aggregates `ablation/results/all_runs.csv` and reproduces the table used in the paper and presentation.

```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd().resolve()
if PROJECT_ROOT.name == 'notebooks':
    PROJECT_ROOT = PROJECT_ROOT.parent

df = pd.read_csv(PROJECT_ROOT / 'ablation' / 'results' / 'all_runs.csv')
df.sort_values('best_val_dice_mean', ascending=False)
```

```python
summary = (
    df.groupby('scenario', as_index=False)['best_val_dice_mean']
      .max()
      .sort_values('best_val_dice_mean', ascending=False)
)
summary
```
